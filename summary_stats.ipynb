```python
import sqlite3
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
conn = sqlite3.Connection("data.db")
data = pd.read_sql("SELECT * FROM prices WHERE ticker IN ('MSFT', 'AAPL')", conn,
                   index_col=["ticker", "date"], parse_dates=['date'])
```

```python
data['adjustment_factor'] = 1 + (data['dividends'] / data['close'])
data['adjustment_factor'] = data.groupby(level=0)['adjustment_factor'].cumprod()

data[['open', 'high', 'low', 'close']] = data[['open', 'high', 'low', 'close']].multiply(data['adjustment_factor'], axis=0)
```

```python
msft_returns = data.loc['MSFT']['close'].pct_change().dropna()
```
